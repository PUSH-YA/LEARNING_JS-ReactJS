```python
# Import the required libraries 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Defining constant values 

G = 10
ms = 10
delta_t = 0.1
```

```python
def DrawSun(): # Define the function 
    fig, ax = plt.subplots()
    drawing_circle = patches.Circle((0,0),1, color = 'r', linestyle = "--", fill = False)
    ax.add_patch(drawing_circle)
    plt.scatter(0,0,color = 'r', marker = 'x')
    return (fig,ax) # to work with draw_orbit
```

```python
DrawSun()
```

We compute a function, $EvalAcc$ that takes in the $x$ and $y$ components of the positions to evaluate the componenets of acceleration due to gravity associated with the position using:
$a(t_i) = \frac{-Gm_s}{r^2(t_{i-1})}\hat{r}(t_{i-1})$

In order to do so, we first take the position vector, $r$, in relation to the sun and turn it into a unit vector based on the equations $\hat{r} = \frac{1}{r}r$, $\hat{r_x} = \frac{r_x}{r}$, $\hat{r_y} = \frac{r_y}{r}$ and then multiply by the appropriate constants given by the above equation(8) to get the x and y components of the acceleration. 

```python
def EvalAcc(rx, ry):  # Define the function with parameters rx and ry
    r_mag2 = rx**2 + ry**2 # Square of the magnitude of vector r for equation 8 which is given as eqn 3
    rx_hat = rx/np.sqrt(r_mag2) # Based on the \hat{r} equations to turn the position vector to unit vector
    ry_hat = ry/np.sqrt(r_mag2) # same as above
    ax = -G * ms * rx_hat / r_mag2 # compute x and y components of acceleration based on equation 8
    ay = -G * ms * ry_hat / r_mag2 # same as above
    return [ax, ay] # return x and y components
```

```python
# Test Case 
EvalAcc(3,4)
```

We will now write a function $UpdateV$ that takes in the previously computed components of velocity, the newly computed components of acceleration, and a time step size. It should return the components of velocity that have been updated using:

$v(t_i) = v(t_{i-1})+a(t_i)\Delta{t}$

In order to do so, we multiply x and y components of acceleration by $\Delta{t}$, we add that to the x and y components of the old velocity to get the new x and y components of the new velocity.

```python
def UpdateV(vxold,vyold,ax,ay,dt): # Define the function with the required parameters
    vx = vxold + ax*dt # compute the formula to find the x component
    vy = vyold + ay*dt # do the same for the y component
    return [vx,vy] # return x and y components
```

```python
# Test Case
UpdateV(3,4,1,2,delta_t)
```

Now we write a function called $UpdateR$ tha takes in the previously computed components of position, the newly componets of position, the newly computed component of velocity, and a time step. The function should return the new components of position calculated using:

$r(t_i) = r(t_{i-1})+v(t_i)\Delta{t}$

In order to do so, we multiply the x and y components of velocity by $\Delta$, we add to the x and y components of the old positions to get the new x and y components of position.

```python
def UpdateR(rxold,ryold,vx,vy,dt):# Define the function with the required parameters
    rx = rxold + vx*dt # compute the formula to find the x component
    ry = ryold+vy*dt # compute the formula to find the y component
    return [rx,ry] # return the x and y components
```

```python
UpdateR(2,3,1,3,delta_t)
```

We now write a function called $GetOrbit$ that takes in components of the initial conditions on the velocity and acceleration as well as a time step. It should return a list of x coordinates and a corresponding list of y coordinates that the planet will be located at.

1. We first use x and y positions to get acceleration by calling $EvalAcc$ 
2. Then, we use x and y components of acceleration in $UpdateV$ to get the new velocity 
3. Then, we use the new velocity in $UpdateR$ to get the new position 
4. The new position will be added to the x and y orbit lists for $GetOrbit$

```python
def GetOrbit(rxi,ryi,vxi,vyi,dt):# Define the function with the required parameters
    rx_list,ry_list = [], [] # Inialize the 2 component lists
    rx_list.append(rxi) # We append initial positions to the list
    ry_list.append(ryi) # Same situation here 
    ax, ay = EvalAcc(rxi,ryi) # We get the acceleration, step 1
    vx, vy = UpdateV(vxi,vyi,ax,ay,dt) # Implement step 2
    rx, ry = UpdateR(rxi,ryi,vx,vy,dt) # Implement step 3
    rx_list.append(rx) # Update list to add the values from the steps above
    ry_list.append(ry) # Same situation here for the y values 
    for i in range(50): # Initialize a for loop to iterate through by implementing steps 1-3
        ax, ay = EvalAcc(rx_list[-1], ry_list[-1]) # Take the most recent position aka last value of the list to get the acceleration
        vx, vy = UpdateV(vx, vy, ax, ay, dt) # implement a recurssion to update velocities based on the old velocities
        rx, ry = UpdateR(rx, ry, vx, vy, dt) # update positions based on the old positions
        rx_list.append(rx) # append these values to the list we created
        ry_list.append(ry) # same situation
    return [rx_list,ry_list]
```

We now draw the orbin that corresponds to the initial conditions $r(0) = [5,0]$ and $v(0) = [0,5]$

```python
def draw_orbit(rx,ry,vx,vy): # Define a function with the given parameters
    fig, ax = DrawSun() # Call DrawSun function to create the
    rx,ry = GetOrbit(rx,ry,vx,vy,delta_t)
    plt.plot(rx,ry)
    plt.xlabel("X component")
    plt.ylabel("Y component")
    plt.title('Orbit')
    ax.set_xlim(-20,20)
    ax.set_ylim(-20,20)

draw_orbit(5,0,0,5)
```

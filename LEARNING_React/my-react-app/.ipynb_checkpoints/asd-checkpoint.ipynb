```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def DrawSun(): # Define the function 
    fig, ax = plt.subplots()
    drawing_circle = patches.Circle((0,0),1, color = 'r', linestyle = "--", fill = False)
    ax.add_patch(drawing_circle)
    plt.scatter(0,0,color = 'r', marker = 'x')
```
